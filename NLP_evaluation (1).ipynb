# Accuracy

```python
acc_df=df_results[df_results['performance metric & model'].str.contains("Accuracy")]
acc_df.head()
```

## Accuraccy validation

```python
acc_df_val=acc_df[acc_df['performance metric & model'].str.contains("val")]
acc_df_val
```
